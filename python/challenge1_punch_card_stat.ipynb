```python
import pandas as pd 
import numpy as np

# To load the JSON data into dataframe
df = pd.read_json('challenge_1_punch_card_stat.json')
```

```python
# rename dataframe title

df.columns = ['Day of week', 'Hour of day', 'Number of commits']
df
```

```python
import matplotlib.pyplot as plt

x = df['Day of week']
y = df['Hour of day']

#enlarge the scale of marker by a power of 1.8
s = df['Number of commits']**1.8
plt.scatter(x,y,s=s, color='r')
plt.xlabel('Day of week')
plt.ylabel('Hour of day (local time)')

plt.xticks([0,1,2,3,4,5,6],['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'])
plt.yticks(range(24),df['Hour of day'])

plt.show()
```
